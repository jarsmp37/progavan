```python
df=df.rename(columns={"age":"Edad","gender":"Genero","smoking_status":"Fumador","bmi":"Indice_masa_corp",
                      'blood_pressure':"Presion_Sanguinea", 'glucose_levels':"Nivel_glucosa", 'condition':"Condicion"})
df
```

```python
#borrar columna
df=df.drop(columns=["id"])
df
```

```python
df.columns
```

```python
for columna in df.columns:
    print(f"La columna {columna} tiene de datos")
    print(df[columna].unique())
    print("")
```

```python
dic1={'male':"Hombre", 'female':"Mujer",'Non-Smoker':"No_fumador", 'Smoker':"Fumador",'Pneumonia':"Neumonia",
        'Diabetic':"Diabetico"}
```

```python
df["Genero"]=df["Genero"].replace(dic1)
df
```

```python
for columna in df.columns:
    df[columna]=df[columna].replace(dic1)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(13,6))
df["Edad"].value_counts().plot(kind="bar",color="blue")
plt.title("Edades")
plt.xlabel("Edad")
plt.ylabel("Conteo")
plt.show()
```

```python
plt.figure(figsize=(13,6))
df["Edad"].value_counts().sort_index().plot(kind="bar",color="blue")
plt.title("Edades")
plt.xlabel("Edad")
plt.ylabel("Conteo")
plt.show()
```

```python
df_filtrado=df[df["Edad"]!=54]
```

```python
df_filtrado=df[~df["Edad"].isin([52,53,54])]
```

```python
plt.figure(figsize=(13,6))
df_filtrado["Edad"].value_counts().sort_index().plot(kind="bar",color="blue")
plt.title("Edades")
plt.xlabel("Edad")
plt.ylabel("Conteo")
plt.show()
```

```python
conteogenero=df["Genero"].value_counts()
plt.pie(conteogenero,labels=conteogenero.index,autopct="%1.1f%%")
plt.show()
```
