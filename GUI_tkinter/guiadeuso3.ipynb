# 1. El Backend: Programación Orientada a Objetos
En esta parte definimos la "plantilla" de los datos. Crearemos una clase llamada `Estudiante`. Cada vez que alguien se registre en la interfaz, el sistema creará un **objeto** único basado en esta clase.

Esto nos permite manipular los datos de forma profesional, guardarlos en listas o bases de datos fácilmente.

```python
class Estudiante:
    def __init__(self, nombre, carrera, semestre):
        self.nombre = nombre
        self.carrera = carrera
        self.semestre = semestre
    
    def obtener_resumen(self):
        return f"Estudiante: {self.nombre} | Carrera: {self.carrera} | Semestre: {self.semestre}"

# Lista global para almacenar los objetos que vayamos creando
base_de_datos = []
```

# 2. El Frontend: Interfaz con Tkinter
Aquí creamos el formulario. El objetivo es que los datos que el usuario escriba en los `Entry` (cuadros de texto) se capturen y se envíen al constructor de nuestra clase `Estudiante`.

**Proceso:**
1. El usuario llena el formulario.
2. Al pulsar "Registrar", se extraen los strings de la UI.
3. Se crea un objeto `nuevo_estudiante = Estudiante(...)`.

```python
import tkinter as tk
from tkinter import messagebox

def registrar_usuario():
    # Extraemos los datos de la interfaz (Frontend)
    nom = entry_nombre.get()
    car = entry_carrera.get()
    sem = entry_semestre.get()
    
    if nom and car and sem:
        # --- AQUÍ OCURRE LA MAGIA DEL BACKEND ---
        # Creamos un objeto con los datos recolectados
        nuevo_estudiante = Estudiante(nom, car, sem)
        
        # Guardamos el objeto en nuestra lista
        base_de_datos.append(nuevo_estudiante)
        
        # Limpiamos los campos y confirmamos
        entry_nombre.delete(0, tk.END)
        entry_carrera.delete(0, tk.END)
        entry_semestre.delete(0, tk.END)
        
        messagebox.showinfo("Éxito", f"Objeto creado:\n{nuevo_estudiante.obtener_resumen()}")
        print(f"Objetos en memoria: {len(base_de_datos)}")
    else:
        messagebox.showwarning("Error", "Por favor llena todos los campos")

# Configuración de la Ventana
root = tk.Tk()
root.title("Registro de Estudiantes")
root.geometry("300x250")

# Widgets de la Interfaz
tk.Label(root, text="Nombre:").pack(pady=5)
entry_nombre = tk.Entry(root)
entry_nombre.pack()

tk.Label(root, text="Carrera:").pack(pady=5)
entry_carrera = tk.Entry(root)
entry_carrera.pack()

tk.Label(root, text="Semestre:").pack(pady=5)
entry_semestre = tk.Entry(root)
entry_semestre.pack()

# Botón que dispara la creación del objeto
btn_registrar = tk.Button(root, text="Registrar Estudiante", command=registrar_usuario)
btn_registrar.pack(pady=20)

root.mainloop()
```
