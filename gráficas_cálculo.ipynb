#Función lambda

```python
def composicion(f, g):
    return lambda x: f(g(x))

def composicion2(f, g):
    return lambda x: g(f(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt
def funcion1(x):
    return ((x+3)** 2)-4
def funcion2(x):
    return -((5-2*x)**0.5)+3
h = composicion(funcion1, funcion2)
x1 = np.linspace(-4, 7, 500)
x2 = np.linspace(-1000, 5/2, 500)
y1 = funcion1(x1)  
y2 = funcion2(x2)
xinter=np.linspace(-4, 3, 500)  
y_compuesto = h(xinter)
  
plt.figure(figsize=(8, 5))
plt.plot(x1, y1, label="f(x)", color='blue')
plt.plot(x2, y2, label="g(x)", color='green')
plt.plot(xinter, y_compuesto, label="f(g(x))", color='red', linestyle='--')
plt.title("Gráfica de las funciones y su composición")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.ylim(-9,125)
plt.xlim(-5,8)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import interact, FloatSlider

def parabola(a, h, k):
    x = np.linspace(-10, 10, 400)
    y = a * (x - h)**2 + k
    return x, y

def plot_parabola(a=1, h=0, k=0):
    x, y = parabola(a, h, k)
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, label=f'y = {a}(x - {h})² + {k}')
    plt.title('Parábola Interactiva')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black', linewidth=0.5)
    plt.axvline(0, color='black', linewidth=0.5)
    plt.grid(True)
    plt.legend()
    plt.ylim(-10, 10)
    plt.xlim(-10, 10)
    plt.show()

interact(plot_parabola, 
         a=FloatSlider(value=1, min=-5, max=5, step=0.1, description='a:'),
         h=FloatSlider(value=0, min=-5, max=5, step=0.1, description='h:'),
         k=FloatSlider(value=0, min=-5, max=5, step=0.1, description='k:'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider

# Definimos la función que queremos graficar
def f(x):
    return x**2

# Definimos el punto en el que queremos evaluar el límite
a = 2
L = f(a)  # El límite es f(a) en este caso

# Creamos la figura y los ejes
fig, ax = plt.subplots()
plt.subplots_adjust(left=0.1, bottom=0.35)

# Creamos el rango de x para la gráfica
x = np.linspace(0, 4, 400)
y = f(x)

line, = plt.plot(x, y, lw=2)


point, = plt.plot(a, L, 'ro')


epsilon_line1, = plt.plot([], [], 'g--', lw=1)
epsilon_line2, = plt.plot([], [], 'g--', lw=1)
delta_line1, = plt.plot([], [], 'b--', lw=1)
delta_line2, = plt.plot([], [], 'b--', lw=1)

ax_epsilon = plt.axes([0.25, 0.15, 0.65, 0.03])
ax_delta = plt.axes([0.25, 0.1, 0.65, 0.03])
epsilon_slider = Slider(ax_epsilon, 'Epsilon', 0.1, 2.0, valinit=1.0)
delta_slider = Slider(ax_delta, 'Delta', 0.1, 1.0, valinit=0.5)

def update(val):
    epsilon = epsilon_slider.val
    delta = delta_slider.val
    
    
    epsilon_line1.set_data([a - delta, a + delta], [L - epsilon, L - epsilon])
    epsilon_line2.set_data([a - delta, a + delta], [L + epsilon, L + epsilon])
    
    
    delta_line1.set_data([a - delta, a - delta], [L - epsilon, L + epsilon])
    delta_line2.set_data([a + delta, a + delta], [L - epsilon, L + epsilon])
    
    
    point.set_data(a, L)
    
    fig.canvas.draw_idle()

epsilon_slider.on_changed(update)
delta_slider.on_changed(update)


plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Definición de Límite con Epsilon y Delta')
plt.grid(True)
plt.show()
```
