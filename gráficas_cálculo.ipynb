## Gráficas

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(-50,10,100)

y=(-(12-9*x)**0.5)-2

plt.figure(figsize=(6,4))

plt.plot(x,y,label="y=2x+3",linewidth=3.8,color="red")
plt.title("Gráfica de una recta")
plt.xlabel("valores de x")
plt.ylabel("valores de y")
plt.axhline(0,color="black",linewidth=1,linestyle="--")
plt.axvline(0,color="black",linewidth=1,linestyle="--")
plt.legend(fontsize=13)

plt.ylim(-25,3)
plt.xlim(-50,20)

plt.show()
```

#Función lambda

```python
def composicion(f, g):
    return lambda x: f(g(x))

def composicion2(f, g):
    return lambda x: g(f(x))
```

```python
import numpy as np
import matplotlib.pyplot as plt
def funcion1(x):
    return ((x+3)** 2)-4
def funcion2(x):
    return -((5-2*x)**0.5)+3
h = composicion(funcion1, funcion2)
x1 = np.linspace(-4, 7, 500)
x2 = np.linspace(-1000, 5/2, 500)
y1 = funcion1(x1)  
y2 = funcion2(x2)
xinter=np.linspace(-4, 3, 500)  
y_compuesto = h(xinter)
  
plt.figure(figsize=(8, 5))
plt.plot(x1, y1, label="f(x)", color='blue')
plt.plot(x2, y2, label="g(x)", color='green')
plt.plot(xinter, y_compuesto, label="f(g(x))", color='red', linestyle='--')
plt.title("Gráfica de las funciones y su composición")
plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
plt.axvline(0, color="black", linewidth=0.5, linestyle="--")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.ylim(-9,125)
plt.xlim(-5,8)
plt.show()
```
