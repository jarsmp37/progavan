```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función y su derivada
def f(x):
    return np.sin(x)  # Ejemplo: función seno

def df(x):
    return np.cos(x)  # Derivada del seno es coseno

# Punto donde aproximamos (a, f(a))
a = np.pi / 4  # Ejemplo: a = π/4 (45 grados)
f_a = f(a)
df_a = df(a)

# Función de la recta tangente (aproximación lineal)
def tangent_line(x):
    return f_a + df_a * (x - a)

# Rango de valores para graficar
x_vals = np.linspace(0, np.pi, 100)  # De 0 a π
y_vals = f(x_vals)
y_tangent = tangent_line(x_vals)

# Graficar
plt.figure(figsize=(10, 6))
plt.plot(x_vals, y_vals, label='Función: $f(x) = \sin(x)$', color='blue')
plt.plot(x_vals, y_tangent, label=f'Aproximación lineal en $x = \pi/4$', color='red', linestyle='--')
plt.scatter(a, f_a, color='green', label=f'Punto de tangencia: $(\pi/4, \sin(\pi/4))$')

plt.title('Aproximación Lineal usando la Derivada')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```

### 🌱 Ejemplo 1: Tasa de cambio del área de un círculo

El radio de un círculo está aumentando a razón de **3 cm/s**.  
¿A qué velocidad está cambiando el **área del círculo** cuando el radio es de **5 cm**?

📌 Datos:

$$
A = \pi r^2
$$

$$
\frac{dr}{dt} = 3 \, \text{cm/s}
$$

Se pide:

$$
\frac{dA}{dt} \quad \text{cuando} \quad r = 5
$$


```python
import sympy as sp
import numpy as np
r=sp.Symbol("r")
A=np.pi*r**2
A_der=sp.diff(A,r)
Da_dt=A_der.subs(r,5)*3
Da_dt
```

### 🚗 Ejemplo 2: Velocidad en un movimiento rectilíneo

La posición de un objeto en movimiento rectilíneo está dada por la función:

$$
s(t) = 2t^3 - 5t^2 + 4t
$$

donde \( s \) está en **metros** y \( t \) en **segundos**.  
¿Cuál es la **velocidad del objeto** cuando \( t = 2 \, \text{s} \)?

📌 Recuerda:

La velocidad es la derivada de la posición:

$$
v(t) = \frac{ds}{dt}
$$

Se pide:

$$
v(2)
$$


```python
import sympy as sp
t=sp.Symbol('t')
s=2*t**3-5*t**2+4*t
v=sp.diff(s,t)
v_en_2=v.subs(t,2)
v_en_2
```

### 🎈 Ejemplo 3: Tasa de cambio del volumen de una esfera

Una esfera se está inflando de manera que su radio crece a razón de:

$$
\frac{dr}{dt} = 2 \, \text{cm/s}
$$

¿Qué tan rápido está cambiando su **volumen** cuando el radio es de **4 cm**?

📌 Fórmula del volumen de una esfera:

$$
V = \frac{4}{3} \pi r^3
$$

Se pide:

$$
\frac{dV}{dt} \quad \text{cuando} \quad r = 4
$$


```python
import sympy as sp
import numpy as np
r=sp.Symbol('r')
V=(4/3)*np.pi*r**3
V_der=sp.diff(V,r)
V_der_sub=V_der.subs(r,4)*2
V_der_sub
```

### 📊 Ejemplo 4: Crecimiento de registros en una base de datos

Supongamos que la cantidad de registros en una base de datos crece según la función:

$$
R(t) = 500 \cdot \ln(t + 1)
$$

donde \( R(t) \) representa el número de registros después de \( t \) días.

¿Cuál es la **tasa de crecimiento de registros** en el día 10?

📌 Se pide:

$$
\frac{dR}{dt} \quad \text{cuando} \quad t = 10
$$


```python
import sympy as sp
t=sp.Symbol("t")
R=500*sp.log(t+1)
R_der=sp.diff(R,t)
R_der_en10=R_der.subs(t,10)
R_der_en10
```
