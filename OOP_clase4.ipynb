## importar entre diferentes archivos

#  ¿Cómo usar un archivo `main` en Python?

Cuando trabajas con Programación Orientada a Objetos, es común que:
- En **un archivo** coloques tus clases.
- En **otro archivo** ejecutes el programa principal.

A esto se le llama usar un **archivo main**.

---
##  1. Crea dos archivos

### Archivo 1: contiene la clase
Guarda tu clase (por ejemplo, `Banco`) en un archivo con nombre:


```python
class Banco:
    def __init__(self,nombre,saldo):
        self.__nombre=nombre
        self.__saldo=saldo

    #Getter
    def nombrecliente(self):
        return self.__nombre
    
    def obtenersaldo(self):
        return self.__saldo
    
    #Setter
    def deposito(self,cant1):
        self.__saldo += cant1
        return f"Tu nuevo saldo es {self.__saldo}"
    
    def retirar(self,cant2):
        if self.__saldo<cant2:
            return "No tienes saldo suficiente"
        else:
            self.__saldo-=cant2
            return f"Tu nuevo saldo es {self.__saldo}"
```

Aquí es donde vas a usar la clase que guardaste en `banco.py`.

---

##  2. Importar la clase en el main
Para poder usar la clase que está en otro archivo, el `main.py` debe **importarla**.  
Esto se hace con la instrucción:



```python
from banco import Banco

cliente1=Banco("Angel",100000)
print(cliente1.nombrecliente())
print(cliente1.obtenersaldo())
print(cliente1.deposito(20000))
print(cliente1.obtenersaldo())
print(cliente1.retirar(60000))
print(cliente1.retirar(70000))
```


Esto le dice a Python:  
> “Trae la clase Banco desde el archivo banco.py”.

---

##  3. Crear objetos y ejecutar el programa desde el main

En el archivo `main.py`:
- Creas los objetos.
- Llamas a los métodos.
- Imprimes los resultados.

Todo lo que quieras que el programa haga al ejecutarse debe estar en este archivo.

---

##  ¿Por qué se usa un archivo main?

- Separa la **lógica** (clases) del **uso** de esas clases.
- Mantiene tu código **ordenado y profesional**.
- Permite reutilizar tus clases en otros programas sin copiarlas.
- Es la forma estándar de trabajar en proyectos más grandes.

---

## ✔️ ¿Cómo lo ejecutas?

Simple:  
Ejecuta **solo** el archivo `main.py`.
