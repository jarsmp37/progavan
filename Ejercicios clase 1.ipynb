# Ejercicio: Crear una clase en Python

## Objetivo
En este ejercicio vas a crear tu propia clase en Python utilizando el concepto de **Programación Orientada a Objetos (POO)**.  
El objetivo es que definas una clase simple, la inicialices con atributos y crees un método que muestre información del objeto.

---

## Instrucciones

### 1. Crea una clase llamada `Animal`
Tu clase debe contener:  nombre, color, patas

### 2. Crea un método llamado `descripcion`
- Este método debe imprimir un mensaje parecido al siguiente:

Soy un perro, tengo color gris y tengo 4 patas

## Comprobación

crea 3 objetos diferentes y compruebalos: 

Ejemplo

# Ejercicio: Clase `Banco` (getters y setters)

## Instrucciones

1. Crea una clase llamada **Banco** con:
   - Atributo privado para el **nombre del cliente**.
   - Atributo privado para el **saldo**.

2. En el constructor (`__init__`):
   - Recibe el nombre y el saldo inicial.
   - Guarda ambos en los atributos privados.

3. Crea **dos getters**:
   - Uno para obtener el nombre.
   - Otro para obtener el saldo.

4. Crea **dos métodos para modificar el saldo**:
   - **Depósito:** suma una cantidad al saldo y devuelve un mensaje con el nuevo saldo.
   - **Retiro:** valida si hay saldo suficiente; si sí, resta la cantidad y devuelve el nuevo saldo; si no, muestra mensaje de “saldo insuficiente”.

5. Crea un objeto de tu clase y:
   - Muestra el nombre del cliente.
   - Muestra el saldo inicial.
   - Realiza un depósito.
   - Realiza un retiro.
   - Muestra los resultados.
