# 🧠 PARTE 1: Fundamentos del Prompt Engineering

## 1.1 ¿Qué es el Prompt Engineering?

**Prompt Engineering NO es simplemente "hablarle a la IA".**

Es una **disciplina técnica** que combina:

- 🔬 Conocimiento del comportamiento de los modelos de lenguaje (LLMs)
- 🏗️ Arquitectura de instrucciones estructuradas y precisas
- 🧪 Metodología experimental: prueba, mide, itera
- 🎯 Ingeniería de contexto, restricciones y ejemplos

### ¿Por qué es una habilidad técnica y no solo "escribir bonito"?

| Lo que hace el usuario promedio | Lo que hace el Prompt Engineer |
|--------------------------------|--------------------------------|
| `"Hazme un login"` | Especifica lenguaje, framework, patrón de autenticación, manejo de errores y tests |
| `"Explícame las listas"` | Define la audiencia, nivel técnico, analogías, ejemplos y formato de salida |
| `"Mejora este código"` | Indica métricas objetivo: complejidad ciclomática, principios SOLID, cobertura de tests |
| `"No funciona"` | Provee stacktrace, versión exacta, entorno de ejecución y comportamiento esperado |

> 💡 **Regla de oro:** Un buen prompt es como una **especificación técnica de software**: cuanto más preciso y estructurado, mejor el resultado.

## 1.2 Los 4 Pilares del Prompt Perfecto

```
┌──────────────────────────────────────────────────────────────┐
│                   ANATOMÍA DE UN PROMPT                       │
│                                                              │
│   [ROL]  +  [CONTEXTO]  +  [INSTRUCCIÓN]  +  [FORMATO]       │
│     👤         📄               📌               🗂️            │
└──────────────────────────────────────────────────────────────┘
```

---

### 👤 Pilar 1: ROL — *¿Quién responde?*

Define el perfil experto que debe adoptar el modelo. El rol activa el "modo experto" y calibra el nivel de respuesta.

```
❌ Sin rol:   "Explica los decoradores de Python"

✅ Con rol:   "Actúa como Senior Python Developer con 10 años de
               experiencia en arquitectura de software. Explica
               los decoradores de Python a un desarrollador Junior."
```

---

### 📄 Pilar 2: CONTEXTO — *¿Cuál es la situación?*

Proporciona el estado actual del problema: código existente, entorno, herramientas, restricciones del negocio.

```
❌ Sin contexto:  "Optimiza esta función"

✅ Con contexto:  "Tengo esta función en Python 3.11, usando FastAPI 0.100,
                   que procesa 10,000 requests por segundo.
                   El cuello de botella detectado con cProfile es en la
                   serialización JSON. El código debe mantenerse compatible
                   con la versión actual de la API."
```

---

### 📌 Pilar 3: INSTRUCCIÓN — *¿Qué hacer exactamente?*

Define la acción con verbos concretos y criterios de éxito medibles.

```
❌ Vaga:    "Haz algo con esto"

✅ Precisa: "Refactoriza aplicando el principio Open/Closed de SOLID.
             Mantén compatibilidad con la API existente.
             Incluye type hints en todos los parámetros y retornos.
             Agrega docstrings en formato Google Style."
```

---

### 🗂️ Pilar 4: FORMATO — *¿Cómo quiero la respuesta?*

Especifica la estructura de salida: JSON, Markdown, código, tabla, lista numerada, etc.

```
❌ Sin formato:  "Dame información de endpoints de la API"

✅ Con formato:  "Responde en JSON con estos campos obligatorios:
                  endpoint, método_HTTP, parámetros_requeridos,
                  ejemplo_request, ejemplo_response, código_error_posible"
```

## 1.3 Las 3 Técnicas Clave

---

### 🔖 Técnica 1: Delimitadores

Los delimitadores **separan visualmente** las partes del prompt. Evitan que el modelo mezcle instrucciones con datos.

**Opción A — Etiquetas XML** *(recomendado para prompts largos y complejos)*
```xml
<rol>Senior Backend Developer especializado en APIs REST con Python</rol>

<contexto>
  Tengo una API en FastAPI que retorna errores 500 de forma intermitente.
  Stack: Python 3.11, FastAPI 0.100, PostgreSQL 15, Redis 7.
</contexto>

<instruccion>
  Analiza las causas más comunes de errores 500 en este stack.
  Proporciona un checklist de diagnóstico ordenado por probabilidad.
</instruccion>

<formato>Lista numerada con sub-ítems. Incluye comandos de diagnóstico ejecutables.</formato>
```

**Opción B — Triple comillas** *(para insertar código o texto dentro del prompt)*
```
Refactoriza el siguiente código sin cambiar su funcionalidad:

"""
def calc(x,y,op):
    if op=="+": return x+y
    if op=="-": return x-y
"""

Aplica: PEP 8, type hints, docstring y manejo de operación inválida.
```

---

### 🔗 Técnica 2: Chain of Thought (CoT) — *Razona antes de responder*

Le indicas al modelo que **piense en voz alta paso a paso** antes de dar la respuesta final. Especialmente poderoso para lógica compleja.

```
❌ Sin CoT:  "¿Cuál es la complejidad temporal de este algoritmo?"

✅ Con CoT:  "Analiza la complejidad temporal de este algoritmo.
              Antes de dar tu respuesta final, razona paso a paso:

              PASO 1: Identifica los loops y su nivel de anidamiento
              PASO 2: Analiza las operaciones dentro de cada loop
              PASO 3: Considera el mejor caso y el peor caso
              PASO 4: Expresa la notación Big-O con su justificación"
```

---

### 🎯 Técnica 3: Few-Shot Prompting — *Enseña con ejemplos*

Proporcionas **ejemplos de entrada/salida** para que el modelo aprenda el patrón exacto que necesitas. Un ejemplo vale más que mil palabras de descripción.

```
Convierte nombres de variables de camelCase a snake_case.

Ejemplo 1:
  Entrada: "getUserData"
  Salida:  "get_user_data"

Ejemplo 2:
  Entrada: "calculateTotalPrice"
  Salida:  "calculate_total_price"

Ejemplo 3:
  Entrada: "isHTTPSEnabled"
  Salida:  "is_https_enabled"

Ahora convierte estas variables:
["fetchHTTPResponse", "parseJSONData", "validateUserInput", "getUserByID"]
```

### ✅ Prompt Pro — Ahora copia y pega ESTE en tu IA:

```
<rol>
Actúa como Software Architect experto en Clean Code,
principios SOLID y reducción de deuda técnica.
</rol>

<contexto>
Este código Python está en producción y procesa datos financieros.
Tiene alta complejidad ciclomática (CC > 10) y viola principios SOLID.
Restricción crítica: no podemos cambiar la firma de la función pública procesar(datos).
</contexto>

<codigo_a_refactorizar>
def procesar(datos):
    resultado = []
    for d in datos:
        if d['tipo'] == 'A':
            if d['valor'] > 100:
                if d['activo'] == True:
                    resultado.append(d['valor'] * 1.15)
                else:
                    resultado.append(d['valor'])
            else:
                resultado.append(d['valor'] * 0.9)
        elif d['tipo'] == 'B':
            if d['valor'] > 50:
                resultado.append(d['valor'] * 1.2)
            else:
                resultado.append(d['valor'] * 0.85)
        else:
            resultado.append(d['valor'])
    return resultado
</codigo_a_refactorizar>

<instruccion>
Razona en voz alta ANTES de escribir código, siguiendo estos pasos:

PASO 1 — ANÁLISIS: Identifica cada violación SOLID con su principio exacto.
PASO 2 — PLAN: Define la estrategia de refactorización y los patrones a aplicar.
PASO 3 — IMPLEMENTACIÓN: Escribe el código refactorizado.
PASO 4 — MÉTRICAS: Compara la complejidad ciclomática antes vs. después.

Restricciones:
- La función procesar(datos) debe mantener la misma firma pública
- Usa Python 3.11 con type hints obligatorios
- Aplica Strategy Pattern para los distintos tipos de procesamiento
</instruccion>

<formato>
Usa encabezados Markdown para cada paso.
El código refactorizado debe estar en un bloque Python al final.
</formato>
```

**¿Qué diferencia notas?** → La IA analiza las violaciones SOLID, justifica sus decisiones, aplica Strategy Pattern y reporta la mejora en complejidad.

# 🏥 DESAFÍO FINAL 2: POO Aplicada — Sistema de Consultorio Médico

> **Objetivo:** Usar Prompt Engineering para diseñar y generar un sistema completo
> orientado a objetos, partiendo de un enunciado de negocio real.

---

## 🩺 El Escenario

El **Dr. Ramírez** necesita digitalizar su consultorio médico. Lleva todo en papel
y quiere un sistema en Python que gestione pacientes, médicos, citas y expedientes clínicos.

### Entidades del sistema

| Entidad | Atributos clave | Comportamientos |
|---------|----------------|------------------|
| **Paciente** | curp, nombre, fecha_nacimiento, tipo_sangre, alergias | calcular_edad(), tiene_alergia() |
| **Medico** | cedula, nombre, especialidad, horario | esta_disponible(), agendar_cita() |
| **Cita** | folio, paciente, medico, fecha_hora, motivo, estado | confirmar(), cancelar(), reprogramar() |
| **ExpedienteClinico** | paciente, historial_citas, diagnosticos | agregar_diagnostico(), obtener_historial() |
| **Consultorio** | nombre, medicos, pacientes, agenda | registrar_paciente(), buscar_paciente(), generar_reporte() |

### Reglas de negocio

- Un médico **NO puede** tener dos citas solapadas en el tiempo
- Las citas deben agendarse con **mínimo 2 horas** de anticipación
- Pacientes **menores de 18 años** requieren tutor registrado
- Estado de la cita: `PENDIENTE → CONFIRMADA → ATENDIDA | CANCELADA` (flujo unidireccional)
- El expediente se **crea automáticamente** al registrar un paciente
- No pueden existir **dos pacientes con el mismo CURP**

```
<rol>
Actúa como Senior Software Engineer especializado en Programación Orientada
a Objetos con Python, diseño de sistemas de dominio y arquitectura de
software para aplicaciones de gestión empresarial.
</rol>

<contexto>
El Dr. Ramírez necesita digitalizar su consultorio médico.
Stack: Python 3.11, sin frameworks externos (solo biblioteca estándar).
El sistema debe ser mantenible y extensible (principios SOLID).
En el futuro se conectará a una base de datos, por lo que se debe separar
la lógica de negocio de la persistencia (patrón Repository).
</contexto>

<reglas_de_negocio>
1. Un médico NO puede tener dos citas solapadas en el tiempo.
2. Las citas deben agendarse con mínimo 2 horas de anticipación.
3. Pacientes menores de 18 años requieren un tutor registrado.
4. Estado de la cita (flujo unidireccional):
   PENDIENTE → CONFIRMADA → ATENDIDA
   PENDIENTE → CANCELADA
   CONFIRMADA → CANCELADA
   (ATENDIDA y CANCELADA son estados finales — no se pueden revertir)
5. El expediente clínico se crea automáticamente al registrar un paciente.
6. No pueden existir dos pacientes con el mismo CURP.
</reglas_de_negocio>

<ejemplo_de_uso_esperado>
# Así debe funcionar el sistema:

consultorio = Consultorio("Consultorio Dr. Ramírez")

medico = Medico(
    cedula="12345678",
    nombre="Dr. Carlos Ramírez",
    especialidad="Medicina General",
    horario={"lunes": ("09:00", "18:00"), "martes": ("09:00", "18:00")}
)
consultorio.registrar_medico(medico)

paciente = Paciente(
    curp="RAML900101HDFMRN09",
    nombre="Luis Ramírez López",
    fecha_nacimiento=date(1990, 1, 1),
    tipo_sangre="O+",
    alergias=["penicilina", "aspirina"]
)
consultorio.registrar_paciente(paciente)  # crea expediente automáticamente

cita = consultorio.agendar_cita(
    paciente=paciente,
    medico=medico,
    fecha_hora=datetime(2025, 6, 15, 10, 0),
    motivo="Revisión general"
)
# Salida esperada: Cita(folio='CITA-0001', estado=EstadoCita.PENDIENTE)

cita.confirmar()        # PENDIENTE → CONFIRMADA
cita.marcar_atendida()  # CONFIRMADA → ATENDIDA
cita.cancelar()         # ERROR: CitaYaFinalizadaError (no se puede revertir)

expediente = consultorio.obtener_expediente(curp="RAML900101HDFMRN09")
print(expediente.obtener_historial())  # Lista completa de citas del paciente
</ejemplo_de_uso_esperado>

<instruccion>
Razona el diseño ANTES de escribir código. Sigue estos pasos en orden:

PASO 1 — DIAGRAMA UML (ASCII):
  Dibuja el diagrama de clases mostrando atributos, métodos
  y relaciones (herencia, composición, asociación, cardinalidad).

PASO 2 — EXCEPCIONES DE DOMINIO:
  Define las excepciones personalizadas necesarias:
  CitaDuplicadaError, PacienteMenorSinTutorError,
  CURPDuplicadoError, CitaYaFinalizadaError, TransicionEstadoInvalidaError

PASO 3 — ENUMERACIONES Y MODELOS:
  - Enum: EstadoCita, TipoSangre, Especialidad
  - Dataclasses para Paciente, Medico, Cita, ExpedienteClinico
  - Usa @property para encapsulamiento
  - Validaciones en __post_init__

PASO 4 — REPOSITORIOS:
  Implementa PacienteRepository y CitaRepository con operaciones CRUD básicas.
  Usa diccionarios internos como almacenamiento en memoria.

PASO 5 — CLASE CONSULTORIO:
  Implementa todos los métodos de negocio usando los repositorios.
  Aplica el patrón Facade para simplificar la interacción del cliente.

PASO 6 — DEMOSTRACIÓN:
  Escribe un bloque main() que demuestre:
  - Registro de 2 médicos con distintas especialidades
  - Registro de 3 pacientes (uno menor de edad sin tutor que genere error)
  - Agendado de 3 citas (una con conflicto de horario que genere error)
  - Transición de estados (incluyendo una transición inválida)
  - Consulta del expediente clínico con historial

PASO 7 — TESTS CON PYTEST:
  Cubre los siguientes casos:
  - test_no_permite_citas_solapadas()
  - test_transicion_estado_invalida_lanza_error()
  - test_edad_calculada_correctamente()
  - test_curp_duplicado_lanza_error()
  - test_expediente_creado_al_registrar_paciente()
</instruccion>

<formato>
Organiza la respuesta con estos encabezados Markdown:
## 1. Diagrama UML (ASCII)
## 2. Excepciones de Dominio
## 3. Enumeraciones
## 4. Modelos (Dataclasses)
## 5. Repositorios
## 6. Clase Consultorio
## 7. Demostración (main)
## 8. Tests con pytest

Todo el código en un único archivo: consultorio_medico.py
Al final, incluye el comando para ejecutar los tests:
pytest consultorio_medico.py -v
</formato>
```
