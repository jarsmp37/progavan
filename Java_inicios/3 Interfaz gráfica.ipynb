## 1. La Regla de Arquitectura: Separar Lógica de Visualización

El error más común de los estudiantes novatos es poner todo un proceso de análisis de datos o resolución de ecuaciones adentro del bloque de código del Botón que lo ejecuta en la Interfaz. **Esto es una muy mala práctica.**

La Interfaz (Vista) solo debe tener tres responsabilidades:
1. Extraer lo que el usuario escribió (Ej: Leer de una caja de texto).
2. Mandar a llamar a una Clase Lógica (Modelo) independiente para que piense.
3. Recibir la respuesta de esa clase e imprimirla en la pantalla.

Veamos cómo es esto conceptualmente. Primero, pensemos en una clase lógica totalmente independiente, como una vieja conocida nuestra, pero construida puramente sin saber nada de ventanas:

```java
// ARCHIVO INDEPENDIENTE: AnalizadorFinanciero.java
// Esta clase es PURAMENTE LÓGICA MÁTEMATICA. No sabe qué es un botón ni una ventana.

public class AnalizadorFinanciero {
    
    // Encapsulamiento de nuestro estado
    private double capitalInicial;
    private double tasaInteres;
    private int meses;

    // El constructor estricto que lo alimenta de la información exterior
    public AnalizadorFinanciero(double capital, double tasa, int tiempo) {
        this.capitalInicial = capital;
        this.tasaInteres = tasa / 100.0; // Convertimos porcentaje a decimal
        this.meses = tiempo;
    }

    // Método que calcula algoritmos duros usando su propia estructura de objetos
    public double calcularRetornoFinal() {
        double capitalFuturo = this.capitalInicial;
        // Aplicamos interés compuesto a través de un ciclo for
        for (int i = 0; i < this.meses; i++) {
            capitalFuturo += (capitalFuturo * this.tasaInteres);
        }
        return capitalFuturo;
    }
    
    // Podríamos tener decenas de métodos de análisis numérico aquí unidos al estado del problema...
}
```

## 3. Manejo de Eventos (`actionPerformed`)

Toda vista es estúpida y muerta hasta que programemos un **Evento** que la despierte. Cuando alguien da clic en un Botón, ocurre un evento que java intercepta.

Para que NetBeans escriba automáticamente el espacio para el código de este evento:
**Acércate a tu botón "Calcular Proyección" en el diseñador visual, y hazle DOBLE CLIC RÁPIDO seguido el uno del otro.**

Serás lanzado de inmediato a la pestaña "Source" (código) donde la pantalla se acomodará directamente dentro de una función con este aspecto: `private void jButton1ActionPerformed(java.awt.event.ActionEvent evt) { ... }`

Todo lo que programes dentro de esas llaves, se resolverá cuando el Botón sea accionado por un ratón informático humano. Allí vincularemos la Vista, con el código Lógico Matemático que elaboramos con anterioridad.

```java
// ALERTA: Este código se inyecta en el bloque del Botón generado automáticamente en tu VentanaFinanzas.java.

private void jButton1ActionPerformed(java.awt.event.ActionEvent evt) {
    try {
        // PASO 1 (Mundo Vista): Recuperar el texto que el usuario insertó usando función .getText()
        String capLeido = txtCapital.getText();
        String tasaLeida = txtTasa.getText();
        String mesesLeidos = txtMeses.getText();
        
        // PASO 1.5 (Procesamiento): Un TextField de Java SIEMPRE devuelve el formato fundamental String (Texto).
        // Un analizador financiero, de Java, no concibe evaluar Strings, sino doubles o ints matemáticos. Hay que convertirlos (Parseo).
        double finalCapital = Double.parseDouble(capLeido);
        double finalTasa = Double.parseDouble(tasaLeida);
        int finalMeses = Integer.parseInt(mesesLeidos);
        
        // ====== PUNTO CRÍTICO DE DISEÑO =======
        // PASO 2 (Mundo Modelo/Lógica): ¡Instanciamos esa clase robusta que creamos al inicio de este doc!
        // En lugar de hacer algoritmos de contabilidad aquí directo mezclado con los TextBox visuales, 
        // Construimos el Objeto matemático con sus parámetros constructores correspondientes:
        AnalizadorFinanciero cerebroApp = new AnalizadorFinanciero(finalCapital, finalTasa, finalMeses);
        
        // PASO 3: Le ordenamos al Cerebro puro que resuelva el trabajo y nos de el double resultante:
        double resultadoMagico = cerebroApp.calcularRetornoFinal();
        
        // PASO 4 (Retorno a Vista): Usamos lo otorgado por la lógica para plasmarlo de vuelta en un Componente Gráfico
        // Utilizaremos la función SetText que muta el texto dinámicamente in-Game, o durante tiempo de uso activo.
        // Redondeamos con String.format a 2 centavos.
        lblResultadoFinal.setText(String.format("Rentabilidad Acumulada: $%.2f", resultadoMagico));
        
    } catch(NumberFormatException ex) {
        // Esto previene que una letra introducida colapse a Java. Si parsean fallidamente letras a Double, entra aquí:
        lblResultadoFinal.setText("Error: Solo se admiten y procesan caracteres de formato Numérico.");
    } 
}
```

## 4. Mostrando Datasets visuales al Estudiante: Uso del Componente `JTable`

En las herramientas para Ciencia de Datos y Machine Learning es constante la representación de marcos tabulares en crudos o "Dataframes". En Java, mostramos tabulados dentro de sus interfaces usando la potentísima pero delicada **Tabla (`JTable`)**.

Una JTable es únicamente la piel, un "Monitor". Todo lo que enseña lo recauda internamente de su cerebro o alma, conocido en programación Java general como Su Modelo O **DefaultTableModel**.

### Instrucciones NetBeans:
1. Ve a la paleta, en "Swing Controls", localiza la **Table**.
2. Arrástrala directamente. Nota que la tabla suele agruparse sola en un `jScrollPane` (Es un panel con barra de scroll para manejar miles de celdillas gigantes si ocurre).
3. Para editar qué columnas predeterminan de base su alma técnica, haz clic derecho encima -> `Table Contents...` -> Ve a pestaña superior llamada `Columns` -> Modifica los nombres Titulo1... por `["ID_Muestra", "Temp_X", "Status"]` respectivamente y da Close.
4. Clic derecho en la tabla -> Change Variable Name -> `miTablaDeDatos`.

```java
// CÓDIGO EJEMPLO (Puede insertarse en algún Botón de acción "Cargar CSV Visual")
private void btnCargarGridVisualActionPerformed(java.awt.event.ActionEvent evt) {
    
    // 1. Esto nos requiere importar javax.swing.table.DefaultTableModel; en la fase Superior de bibliotecas del archivo IDE.
    // Accedemos de emergencia a el interior o Modelo cerebral nativo mediante casteo.
    javax.swing.table.DefaultTableModel tableModelVirtual = (javax.swing.table.DefaultTableModel) miTablaDeDatos.getModel();
    
    // 2. Antes de volcar información, limpiar cualquier hilera arcaica pasada si existió.
    tableModelVirtual.setRowCount(0);
    
    // 3. Suponemos que mediante lógica o lectura CSV poseemos una Matriz Estándar de Java:
     Object[][] subDatasetHipotetico = {
         {1, 35.8, "ESTABLE"},
         {2, 39.4, "CALENTANDO"},
         {3, 44.0, "ALERTA"} 
     };
    
    // 4. Volcamos la información al Modelo de la Interface usando For...
    for(int j = 0; j < subDatasetHipotetico.length; j++) {
         // Agregamos las renglones iterada uno por uno en una instrucción nativa general
         tableModelVirtual.addRow(subDatasetHipotetico[j]);
    }
}
```

## 6. Integración Total: Proyecto Final Analítico Visual (Clase Simuladora Completa)

Culminaremos con un ejercicio que enlaza Polimorfismo / Estructuras POO completas / Interfaces y Datasets limpios. Generaremos una ventana con un Desplegable, Botón de arranque y una Grilla de Datos visual interactuando.

### **CÓDIGO LÓGICO Y DE SOPORTE POO (Crea esto en un Archivo Puro Independiente)**


```java
import java.util.Random;

// Objeto Autónomo Lógico Puro (Sin conocimiento Visual)
public class SimGridGenerator {
    private String tipoDistribucion;
    private int totalCargas;

    public SimGridGenerator(String dropDownTypeSelected, int cantVol) {
        this.tipoDistribucion = dropDownTypeSelected;
        this.totalCargas = cantVol;
    }

    public Object[][] procesarBatchGeneracional() {
        // Genera Matriz compatible con Swing visual con las dimensiones necesarias y 3 Columnas Estáticas
        Object[][] matrizVirtual = new Object[this.totalCargas][3];
        Random gaussEngine = new Random();
        
        for (int p=0; p < this.totalCargas; p++){
            matrizVirtual[p][0] = "TRX_B_" + (p+1); // Col 0 Clave Identificadora
            
            if(this.tipoDistribucion.equals("Uniforme")) {
                 matrizVirtual[p][1] = Math.round(gaussEngine.nextDouble() * 100);
            } else { 
                 // Usamos math puro de java sin librerías de python
                 matrizVirtual[p][1] = Math.round(gaussEngine.nextGaussian() * 35 + 100); 
            }
            // Logramos categorizar columna Estado Lógicamente
            double valRevisar = Double.parseDouble(matrizVirtual[p][1].toString());
            matrizVirtual[p][2] = (valRevisar > 80.0) ? "RIESGO" : "CONTROLADO";
        }
        return matrizVirtual;
    }
}
```
