# Clase 7: POO I - El Paradigma de Objetos, Clases e Instancias

Bienvenidos al siguiente nivel de la programación. Hasta ahora, hemos trabajado bajo un paradigma llamado **Programación Estructurada o Procedural**, donde el código se lee de arriba hacia abajo y las funciones operan sobre datos separados. 

Cuando los sistemas crecen enormemente (por ejemplo, en el desarrollo del núcleo de Apache Spark, Hadoop, o grandes motores de bases de datos), manejar miles de variables sueltas se vuelve imposible. Aquí es donde entra la **Programación Orientada a Objetos (POO)**.

La POO busca imitar el mundo real. En el mundo real, no tenemos "variables lógicas", tenemos **Objetos**. Un auto, un usuario, una conexión a una base de datos o un modelo de regresión son "objetos". Cada objeto tiene:
1. **Características (Estado):** En Java, a esto le llamamos **Atributos** (variables dentro del objeto).
2. **Comportamientos (Acciones):** En Java, a esto le llamamos **Métodos** (funciones dentro del objeto).

## 1. La Clase vs. El Objeto

> **Regla de oro:** Una **Clase** es el *plano o molde* arquitectónico. El **Objeto** (o *instancia*) es la casa construida a partir de ese plano.

Puedes usar la *clase* `Automovil` para construir mil *objetos* de tipo auto. Cada auto tendrá su propio color, velocidad y dueño, pero todos comparten la misma estructura dictada por el plano.

En nuestro contexto de Ciencia de Datos asimilemos esto: Una fila de tu Dataset dejará de ser múltiples variables (id, nombre, edad) y se convertirá en un **solo objeto** lógico que contiene todas esas variables juntas.

## 2. Abstracción

La abstracción es la capacidad de ignorar los detalles innecesarios y enfocarnos solo en lo que nos importa para nuestro programa. Si hacemos una clase `Estudiante`, no nos importa de qué material están hechos sus zapatos o el color de sus calcetines, solo nos importa su `matricula`, `promedio` y `carrera`.

## 3. Creando Nuestra Primera Clase y Objetos

En Apache NetBeans, un proyecto clásico de POO suele tener varios archivos `.java` en la misma carpeta (paquete). Uno será nuestro molde (la clase conceptual), y otro archivo separado tendrá el famoso `public static void main` donde usaremos ese molde.

### Paso 1: Crear la Clase Molde (Archivo: `ModeloAnalitico.java`)

```java
// Notarás que este archivo NO tiene public static void main.
// Este archivo es puramente un molde conceptual.
public class ModeloAnalitico {
    
    // 1. ATRIBUTOS (El estado del objeto)
    // Se definen directamente dentro de la clase, afuera de cualquier método.
    String nombreAlgoritmo;
    double precision;
    boolean estaEntrenado;

    // 2. CONSTRUCTOR
    // Es un método especial que "construye" o inicializa el objeto en la memoria (RAM)
    // Obligatoriamente DEBE llamarse EXACTAMENTE igual que la clase y no tiene la palabra "void".
    public ModeloAnalitico(String npNombre, double npPrecisionInicial) {
        // Asignamos los valores que nos pasan por paréntesis a nuestros atributos
        nombreAlgoritmo = npNombre;
        precision = npPrecisionInicial;
        estaEntrenado = false; // Por defecto nace sin estar entrenado
    }

    // 3. MÉTODOS (El comportamiento)
    // Acciones que el objeto sabe hacer consigo mismo.
    public void entrenarModelo() {
        System.out.println("Iniciando fase de entrenamiento para " + nombreAlgoritmo + "...");
        // Simulamos un proceso...
        precision = precision + 0.15; // Su precisión sube al entrenar
        estaEntrenado = true;
        System.out.println("Entrenamiento completado exitosamente.");
    }

    public void mostrarDesempenio() {
        System.out.println("--- Reporte del Modelo ---");
        System.out.println("Algoritmo: " + nombreAlgoritmo);
        System.out.println("Precisión actual: " + precision);
        System.out.println("Estado de entrenamiento: " + estaEntrenado);
    }
}
```

### Paso 2: Usar la Clase (Archivo principal: `MainPOO.java`)

En este archivo, sí tendremos nuestro `main`, donde "instanciaremos" (crearemos) los objetos.

```java
public class MainPOO {
    public static void main(String[] args) {
        
        // Vamos a instanciar (fabricar) nuestro primer objeto.
        // TIPO_DE_DATO nombreVariable = new CONSTRUCTOR(valores...);
        ModeloAnalitico modelo1 = new ModeloAnalitico("Regresion Logistica", 0.50);
        ModeloAnalitico modelo2 = new ModeloAnalitico("Arbol de Decisiones", 0.65);
        
        // Cada objeto vive en su propio de espacio de memoria de forma independiente
        // Si entreno al modelo1, el modelo2 no se ve afectado.
        
        System.out.println("--- Estado INICIAL ---");
        modelo1.mostrarDesempenio();
        modelo2.mostrarDesempenio();
        
        System.out.println("\n--- Ejecutando Comportamientos ---");
        // Utilizamos el punto (.) para acceder a los métodos del objeto
        modelo1.entrenarModelo(); 
        
        System.out.println("\n--- Estado FINAL ---");
        modelo1.mostrarDesempenio();
        modelo2.mostrarDesempenio(); // Este sigue en 0.65 y false, porque nunca lo entrenamos
    }
}
```

# Clase 8: POO II - Encapsulamiento, Getters, Setters y 'this'

Uno de los cuatro pilares fundamentales de la Programación Orientada a Objetos es el **Encapsulamiento**. 

En la clase anterior, hicimos nuestros atributos de esta forma: `double precision;`. Esta declaración carece de un *modificador de acceso*, lo que en la vida real es un enorme problema de ingeniería y seguridad informática. Si otro programador toma tu objeto y hace `modelo1.precision = 9000.0;`, alterará arbitraria y directamente los datos críticos de la aplicación, ocasionando que los cálculos matemáticos exploten sin filtro.

El encapsulamiento dicta que **los atributos (estado) de un objeto deben estar a salvo y ocultos del exterior**. Nadie fuera de la clase debe modificarlos directamente.

## 1. Modificadores de Acceso
Son palabras reservadas que ponemos antes de declarar un atributo o método:
* `public`: Accesible desde cualquier parte del proyecto. 
* `private`: Accesible **ÚNICAMENTE** desde dentro de las llaves de la clase a la que pertenecen. ¡Son invisibles para las demás clases!
* `protected`: Lo veremos en el tema de Herencia.

> **Regla de oro del Encapsulamiento:** En cualquier desarrollo real de Java, los atributos se definen como `private` casi en el 100% de los casos.

## 2. Getters y Setters
Si los atributos son privados, ¿cómo leo su valor (`get`) o modifico su valor (`set`) en un futuro? 
Utilizando **Métodos Públicos** que actúen como "porteros" de nuestra información.

* **Getter (Obtenedor):** Un método con `return` que solo sirve para devolver el atributo encapsulado. No puede alterar los datos, solo leerlos.
* **Setter (Establecedor):** Un método `void` que sirve para sobrescribir o asignar un nuevo valor a un atributo. Aquí reside el poder: ¡Puedes poner **condicionales `if`** dentro de los Setters para proteger al objeto de datos absurdos!

## 3. La palabra clave `this`
Es normal de pronto quedarnos sin imaginación para pensar en nombres de parámetros. Cuando el parámetro que pide un método se llama **exactamente igual** que el atributo interno de la clase, Java se confunde.
La palabra clave `this` significa *"Me refiero al de esta clase (el atributo), no al del argumento temporal"*.

```java
public class RegistroDatos {
    private String id;
    
    // Al haber ambigüedad de nombres(id y id), 'this' clarifica al compilador
    public RegistroDatos(String id) {
        this.id = id; 
    }
}
```

## 4. Ejemplo Integrador de Archivos Seguros

### Archivo: `UsuarioBaseDatos.java`
```java
public class UsuarioBaseDatos {
    
    // 1. ATRIBUTOS PRIVADOS (Nadie externo a esta clase puede verlos ni modificarlos)
    private String username;
    private String password;
    private int nivelAcceso;

    // 2. CONSTRUCTOR (Es público para que podamos instanciar el objeto)
    public UsuarioBaseDatos(String username, String password, int nivelAcceso) {
        this.username = username;
        this.password = password;
        
        // Ya desde la construcción aplicamos filtros de seguridad
        if (nivelAcceso < 1) {
            this.nivelAcceso = 1; // Asignamos nivel mínimo preventivo
        } else {
            this.nivelAcceso = nivelAcceso;
        }
    }

    // 3. GETTERS (Métodos de sólo lectura)
    // El usuario es de acceso libre para lectura.
    public String getUsername() {
        return this.username;
    }

    public int getNivelAcceso() {
        return this.nivelAcceso;
    }
    
    // Fíjate que a propósito NO creamos un getPassword(). 
    // Por seguridad en este sistema dictaminamos que la contraseña nunca puede ser leída, solo validada o cambiada.

    // 4. SETTERS (Métodos de escritura controlada)
    // Supongamos que queremos cambiar el nivel de acceso por una promoción.
    public void setNivelAcceso(int nuevoNivel) {
        // En vez de modificación directa, el Setter evalúa si en efecto la petición es lógica
        if (nuevoNivel >= 1 && nuevoNivel <= 5) {
            this.nivelAcceso = nuevoNivel;
            System.out.println("Nivel de acceso actualizado correctamente a: " + this.nivelAcceso);
        } else {
            System.err.println("ERROR: Nivel de acceso denegado. Se mantienen los privilegios previos.");
        }
    }

    // Setter para la contraseña, con validaciones de longitud
    public void setPassword(String passwordActual, String nuevaPassword) {
        if(this.password.equals(passwordActual)) {      // .equals sirve para comparar Textos (Strings) en Java!
            if(nuevaPassword.length() >= 8) {
                this.password = nuevaPassword;
                System.out.println("Contraseña cambiada con éxito.");
            } else {
                System.err.println("ERROR: La nueva contraseña debe tener 8 o más caracteres.");
            }
        } else {
            System.err.println("CREDENCIALES INVÁLIDAS: Contraseña actual incorrecta.");
        }
    }
}
```

### Archivo con `main`: `EntornoDeEjecucion.java`
```java
public class EntornoDeEjecucion {
    public static void main(String[] args) {
        UsuarioBaseDatos dbaAdmin = new UsuarioBaseDatos("root_admin", "admin123", 5);
        
        // Esto lanzaría un SUPER ERROR ROJO DE COMPILACIÓN:
        // dbaAdmin.password = "hakearCuenta1";  <--- Está privada. Java lo evita protegiendo nuestro sistema
        
        // Uso legítimo del Getter
        System.out.println("Usuario activo: " + dbaAdmin.getUsername());
        
        // Intento fallido de Setter por violar lógica de negocio
        dbaAdmin.setNivelAcceso(-10); 
        
        // Intento exitoso de Setter
        dbaAdmin.setPassword("admin123", "RootSeguro4455"); 
    }
}
```

# Clase 9: POO III - Herencia, Super y Sobrescritura

Imagina construir un sistema de un Banco Central. Tienen "Cuentas de Ahorro", "Cuentas Corrientes" y "Tarjetas de Crédito". Todas ellas comparten muchas lógicas: tienen Titular, tienen un Saldo, y requieren sistemas de Autenticación. Escribir este código en tres clases diferentes repitiendo línea por línea es desastroso al momento de dar mantenimiento. 

Aquí brilla otro pilar de la POO: la **Herencia**. Consiste en crear una **Superclase (Padre)** genérica y generalizada, a la cual le anexamos **Subclases (Hijas)** especializadas. La clase Hija hereda gratis absolutamente todos los métodos y los atributos de la clase Padre, permitiéndole reutilizar su código y expandirlo con nuevos elementos que solo le importan a su nicho.

La palabra clave en Java para lograr este vínculo genético es `extends`.

## 1. La Superclase (Molde Genérico)

Vamos a armar un esquema simple utilizado en Machine Learning. Tendremos una gran Superclase de la cual partir llamda `AlgoritmoIA`.

### Archivo: `AlgoritmoIA.java` (El Padre)
```java
public class AlgoritmoIA {
    
    // El modificador 'protected' es el primo hermano de 'private'.
    // protected significa: "Secreto para el exterior, pero mi familia(clases Hijas) SÍ pueden tener acceso a esto".
    protected String nombreAlgoritmo;
    protected int tiempoProcesamientoSegundos;

    // Constructor del padre
    public AlgoritmoIA(String nombreAlgoritmo, int tiempoSegundos) {
        this.nombreAlgoritmo = nombreAlgoritmo;
        this.tiempoProcesamientoSegundos = tiempoSegundos;
    }

    // Método general que todo algoritmo de IA tendrá
    public void ejecutarPrediccion() {
        System.out.println("El algoritmo Genérico está evaluando los umbrales de predicción.");
    }
}
```

## 2. La Subclase (El hijo Especializado - `extends`)

Ahora necesitamos hacer un algoritmo más conciso, una "Red Neuronal". Tiene capas ocultas (algo exclusivo de esa tecnología) pero es por definición también un Algoritmo de IA, por lo tanto hereda lo genérico.

### Archivo: `RedNeuronal.java` (La Hija)
```java
// extends indica que RedNeuronal 'es un tipo de' AlgoritmoIA y hereda su código interno
public class RedNeuronal extends AlgoritmoIA {

    // Atributos exclusivos (nicho) de la hija
    private int cantidadCapasOcultas;

    // CONSTRUCTOR DE LA HIJA
    public RedNeuronal(String nombreAlgoritmo, int tiempoSegundos, int capasOcultas) {
        
        // IMPORTANTE: Al heredar, estás OBLIGADO a llamar al Constructor del Padre primero.
        // Se hace usando obligatoriamente en la primera linea la directiva especial 'super()'
        super(nombreAlgoritmo, tiempoSegundos); 
        
        // Una vez que el padre se inicializó a sí mismo, inicializo las variables propias de la hija
        this.cantidadCapasOcultas = capasOcultas;
    }

    // Método exclusivo de la hija
    public void calibrarPesos() {
        System.out.println("Afinando backpropagation para " + this.cantidadCapasOcultas + " capas...");
        // Fíjate que pudimos imprimir cantidadCapasOcultas... pero también podemos usar 'nombreAlgoritmo'.
        // Y jamás declaramos a nombreAlgoritmo en esta clase.  ¡Viene heredado gratuitamente del padre como protected!
        System.out.println("Modelo " + nombreAlgoritmo + " está siendo calibrado internamente.");
    }
    
    // SOBRESCRITURA (Override)
    // El método 'ejecutarPrediccion' ya existía en el Padre. Pero la red neuronal no predice de forma genérica.
    // Nosotros "Pisamos" o Sobrescribimos el diseño de nuestro papá, para que actúe de forma personalizada aquí.
    @Override
    public void ejecutarPrediccion() {
        // Le indicamos al compilador con @Override (una anotación pura) nuestra intención técnica. 
        System.out.println("PREDICCIÓN RED NEURONAL: Multiplicando las matrices por la entropía cruzada para arrojar decimales de probabilidad final.");
    }
}
```

## 3. Comprobando en Consola (Main)

### Archivo: `MainHerencia.java`
```java
public class MainHerencia {
    public static void main(String[] args) {
        
        System.out.println("--- PRUEBA 1 : SUPERCLASE O PADRE ---");
        AlgoritmoIA simplePadre = new AlgoritmoIA("Regresion Lineal OLS", 2);
        simplePadre.ejecutarPrediccion();
        // simplePadre.calibrarPesos(); // ESTO FALLA. El padre es ignorante del método que se inventó su hijo, no sabe hacerlo.
        
        System.out.println("\n--- PRUEBA 2 : SUBCLASE O HIJO ---");
        RedNeuronal complexHijo = new RedNeuronal("Perceptron Multicapa", 4500, 15);
        
        // ¡Magia! Puede ejecutar su método super calibrado porque lo sobrescribimos con @Override
        complexHijo.ejecutarPrediccion(); 
        
        // Además, puede ejecutar libremente sus métodos y código de nicho exclusivos.
        complexHijo.calibrarPesos();     
    }
}
```

# Clase 10: POO IV - Polimorfismo y Clases Abstractas

Hoy unimos todos los puntos de la Programación Orientada a Objetos en su aplicación de más alto nivel para arquitecturas de computación reales. Ya conocemos y usamos las Clases, Objetos, Encapsulamiento, y Herencia. Hoy entraremos en un concepto sofisticado: El **Polimorfismo**.

**Poli** = muchas; **Morfismo** = formas. El polimorfismo es la habilidad en Java de que una Superclase (Padre) sirva de "paraguas referencial" para contener y operar mútliples clases Hijas muy distintas sin saber exactamente a cúal hija se está llamando bajo el sol.

## 1. El Potencial del Polimorfismo en Arreglos

Imagina que estás gestionando una granja grande de servidores en la nube. Tienes `ServidorWeb`, `ServidorBaseDeDatos`, `ServidorDNS` (que son clases Hijas que heredan de un mismo Padre genérico `ServidorGeneral`).  ¿Cómo apagas mil servidores dispares juntos? ¡Crear tres vectores distintos e iterados es un suplicio!

Gracias al polimorfismo, una variable de tipo SUPERCLASE puede guardar en sus entrañas un objeto instanciado HIPERESPECIALIZADO (clase Hija).

```java
// Válido por Herencia y Polimorfismo. 
// "Un ServidorWeb ES UN ServidorGeneral (superclase)".
ServidorGeneral s1 = new ServidorWeb(); 
ServidorGeneral s2 = new ServidorBaseDeDatos();

// Y lo más impresionante: ¡ARREGLOS POLIMÓRFICOS! 
// Este array unificado puede contener a cualquier clase que haya sido heredada del molde padre.
ServidorGeneral[] miGranja = new ServidorGeneral[3];
miGranja[0] = new ServidorWeb();
miGranja[1] = new ServidorBaseDeDatos();
miGranja[2] = new ServidorBaseDeDatos();

// En un ciclo for, podemos llamar a una funcion general "ejecutarDiagnostico()"
// ¡Magia! A pesar de que todos están referenciados de manera "Padre", cada uno se comportará individualmente 
// con la propia sobrescritura (@Override) que la hija tenga implementada.
for(int i = 0; i < miGranja.length; i++){
    miGranja[i].ejecutarDiagnostico(); 
}
```

## 2. Clases Abstractas (`abstract`)

En el ejemplo anterior creamos instancias de la superclase y de sus hijas. Sin embargo, en la ingeniería pura, usualmente la Superclase Padre es tan teórica y filosófica, que instanciar directamente un padre causa un error arquitectural.

Piensa en los Mamíferos. Un Perro existe. Un Gato existe, y ambas heredan de `Mamifero`. Pero, ¿cómo luce un `Mamifero` neutro e instanciado?. No puedes ir por la vida señalando a un especimen y diciendo "Mira, ese ser es un mamífero puro y generalizado". Todo organismo físico termina siendo una raza, un espécimen final dictado en el árbol genético (un murciélago, un humano...).

Para corregir esto usamos la palabra reservada `abstract`. Al ponerle `abstract` a una clase (antes de la palabra `class`), fuerzas a que este Molde solo sirva para ser heredado. **Jamás alguien podrá usar el comando `new` directamente sobre él.**

Las clases abstractas pueden contener **Métodos Abstractos**, los cuales son comandos que NO TIENEN LLAVES NI CÓDIGO `{}`. Solo obligan contractualmente y amenzan estricto a que las clases Hijas implementen por la fuerza dicho método. 

## 3. Ejemplo Documentado: Pipelines ETL

Veamos cómo en Ciencia de Datos usamos clases abstractas para crear conectores genéricos.

### Archivo Padre Abstracto: `ConectorFuenteAbstracto.java`
```java
// Añadimos 'abstract' a la clase
public abstract class ConectorFuenteAbstracto {
    
    protected String stringConexion;

    public ConectorFuenteAbstracto(String connStr) {
        this.stringConexion = connStr;
    }
    
    // Este método sí tiene cuerpo (llaves), la descendencia lo hereda tal cual.
    public void validacionPingSeguridad() {
        System.out.println("Realizando ping a la URL o IP segura: " + stringConexion + " ... Ok.");
    }
    
    // MÉTODO ABSTRACTO. Nota cómo acaba en punto y coma ';' en lugar de abrir llaves.
    // Esto es un CONTRATO. El Padre dice: "Todo lo que se considere un Conector, está completamente obligado por el IDE a tener una forma de extraer los datos". Sin excusas.
    public abstract void extraerDatosRaw(); 
}
```

### Clase Hija Especializada (PostgreSQL): `ConectorRelacionalSQL.java`
```java
public class ConectorRelacionalSQL extends ConectorFuenteAbstracto {

    public ConectorRelacionalSQL(String laCadena) {
        super(laCadena); // Mandamos la info vital al padre
    }
    
    // El IDE (Apache NetBeans) literalmente marcará todo este archivo en línea roja de error
    // si intentas omitir tu parte contractual del método abastracto. Estamos "obligados" a hacerle su overriding.
    
    @Override
    public void extraerDatosRaw() {
        System.out.println("-> Ejecutando consulta SQL a Tablas ('SELECT * FROM...'). Datos recopilados en Memoria.");
    }
}
```

### Clase Hija Especializada (NoSQL / Hadoop): `ConectorDocumentalMongoDB.java`
```java
public class ConectorDocumentalMongoDB extends ConectorFuenteAbstracto {

    public ConectorDocumentalMongoDB(String laCadena) {
        super(laCadena); 
    }

    @Override
    public void extraerDatosRaw() {
         System.out.println("-> Extrayendo miles de colecciones documentales JSON directamente desde los Clusters masivos de BigData.");
    }
}
```

### Archivo Final con Bucle Polimórfico: `TuberiasETL.java`
```java
public class TuberiasETL {
    public static void main(String[] args) {
        
        // ¡Magia de la Abstracción!
        // No podemos hacer: ConectorFuenteAbstracto error = new ConectorFuenteAbstracto(); // BLOQUEADO, Es Abstracta
        
        // Polimorfismo activo mediante Arreglo:
        ConectorFuenteAbstracto[] motores = new ConectorFuenteAbstracto[2];
        
        // Agrupando dos tecnologías increíblemente dispares bajo su linaje Padre unificado
        motores[0] = new ConectorRelacionalSQL("jdbc:postgresql://localhost:5432/ventas");
        motores[1] = new ConectorDocumentalMongoDB("mongodb://cluster-0:27017/");
        
        System.out.println("====== INICIANDO PIPELINE ANALÍTICO MULTI-TECNOLOGÍA ======");
        
        // Automatización extrema en una sola pasada usando bucles for elementales
        for (int i = 0; i < motores.length; i++) {
            motores[i].validacionPingSeguridad(); // Llamando al método clásico compartido heredado (El PING)
            motores[i].extraerDatosRaw(); // ¡Polimorfismo absoluto! El de índice 0 arrojará SQL e interpretará el for interno a través de sí, el 1 rebotara internamente hacia JSON NoSQL. 
            System.out.println("--- Siguiente Conector ---");
        }
    }
}
```
