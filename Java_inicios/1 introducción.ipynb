¡Bienvenidos al curso básico de Java! Este curso está diseñado específicamente para estudiantes de Ciencias de la Computación e Ingeniería en Ciencia de Datos. Nuestro objetivo es que aprendan las bases sólidas de programación utilizando un lenguaje fuertemente tipado y compilado.

A diferencia de Python (que interpretas en Jupyter Notebooks), Java necesita compilarse antes de ejecutarse. Es el motor detrás de grandes herramientas de procesamiento de datos como Apache Hadoop y Apache Spark.

## 1. ¿Por qué Java?
Java utiliza la filosofía **"Escribir una vez, ejecutar en cualquier lugar"** gracias a la JVM (Java Virtual Machine). Su tipado estricto ayuda a predecir la memoria y a evitar errores en tiempo de ejecución, una destreza crucial en el diseño de arquitecturas de software y pipelines de datos.


## 2. Estructura Fundamental de un Programa
En Java, **todo** debe ir dentro de una "clase". No puedes escribir código suelto como en Python.

> **Nota para Apache NetBeans:** 
> Cuando creas un proyecto, normalmente se genera un archivo base (por ejemplo, `Main.java`). **El nombre de tu archivo público debe coincidir exactamente con el nombre de la clase pública**. 
> Si tu archivo se llama `Main.java`, la primera línea de código debe ser `public class Main`.

### El Método `main`
Para que Java sepa por dónde empezar a ejecutar tu programa, necesitas un punto de entrada obligatorio llamado método `main`.


```java
// Todo programa en Java comienza con una clase public
public class Main {
    
    // Este es el método principal, el "motor" de arranque del programa.
    public static void main(String[] args) {
        // Tu código va aquí dentro
        System.out.println("¡Hola Mundo, listos para programar!");
    }
}
```


## 3. Imprimir en Consola

## 4. Variables y Tipos de Datos Primitivos
Para trabajar con datos estadísticos u operaciones lógicas, necesitas almacenar información en la memoria. A diferencia de Python, en Java debes **declarar explícitamente el tipo de dato** antes de usarlo.

* `boolean`: Valores lógicos verdaderos o falsos (`true` o `false`).
* `char`: Un único carácter encerrado en comillas simples. (Ej. 'A', 'x')

> **Nota sobre Cadenas de Texto:** `String` es un objeto (una clase de Java), no un tipo primitivo. Por eso se escribe con **S** mayúscula. Las cadenas usan comillas dobles `" "`.

## 4. Entrada de Datos Básica con `Scanner`
Para que tu programa no tenga datos "duros" (hardcoded), necesitamos pedirle información al usuario por la terminal. Para ello usamos la clase `Scanner`. 
*Nota: Debes importar la librería de Scanner en la parte más alta del archivo.*

```java
import java.util.Scanner; // <-- IMPORTANTE: Va arriba de la declaración de la clase

public class OperacionesBasicas {
    public static void main(String[] args) {
        
        // 1. Instanciamos el Scanner, diciéndole que lea desde el teclado (System.in)
        Scanner lector = new Scanner(System.in);
        
        System.out.print("Ingresa tu edad: "); 
        // Usamos print() en lugar de println() para que escriba en la misma línea
        int edad = lector.nextInt(); // Lee un número entero de la terminal
        
        // ¡Importante! Si vas a leer un texto después de un número, debes limpiar el "buffer"
        // lector.nextLine(); 
        
        System.out.println("Tu edad es: " + edad);
        
        // Es una buena práctica cerrar el scanner cuando dejas de ocuparlo
        lector.close();
    }
}
```
Otros métodos del Scanner:

* `lector.nextLine()` -> Lee todo un texto con espacios.
* `lector.next()` -> Lee solo una palabra.

# Clase 3: Estructuras de Control de Flujo (Condicionales)

Hacer cálculos y leer datos está bien, pero los programas adquieren "inteligencia" cuando pueden tomar **decisiones** en base a sus datos. En Ciencia de Datos o Ciencias de la Computación, las reglas de decisión y los árboles de decisión son fundamentales.

Para esto en Java contamos con las sentencias de control `if`, `else if`, `else` y la sentencia `switch`.

## 1. Sentencia `if - else`
Esta estructura permite ejecutar un bloque de código SI Y SOLO SI una condición es verdadera (resultado `true`). Si es `false`, entonces ejecutará el bloque contenido dentro del `else`.

```java
int edad = 18;

// La condición obligatoriamente va entre paréntesis ()
// Las llaves { } definen el "ámbito" o el bloque del código. (En vez de indentaciones como en Python)
if (edad >= 18) {
    System.out.println("El algoritmo determina que eres MAYOR de edad.");
} else {
    System.out.println("El algoritmo determina que eres MENOR de edad.");
}
```

## 2. Decisiones Múltiples con `else if`
El mundo rara vez es un simple "blanco o negro". Cuando requieres evaluar condiciones en cascada, usa `else if`. El programa ejecutará la primera condición que devuelva `true` y omitirá el resto.

```java
double probabilidadCarga = 0.85;

if (probabilidadCarga >= 0.90) {
    System.out.println("Altamente probable.");
} else if (probabilidadCarga >= 0.50) {
    // Si entró aquí, ya sabemos que NO es >= 0.90
    System.out.println("Probabilidad media.");
} else {
    System.out.println("Baja probabilidad.");
}
```

## 3. Estructura `switch`
Es una alternativa elegante a una larga cadena de `if - else if` cuando vas a comparar en igualdad y una misma variable frente a múltiples constantes, como opciones de un menú.

```java
int opcion_menu = 2;

switch (opcion_menu) {
    case 1:
        System.out.println("Iniciando ETL...");
        break; // IMPORTANTE: EL break le dice a Java que se salga del switch, sino ejecutará todo lo de abajo
    case 2:
        System.out.println("Filtrando base de datos...");
        break;
    case 3:
        System.out.println("Exportando a CSV...");
        break;
    default:
        // default es el equivalente a "else", es el caso de seguridad.
        System.out.println("Error 404: Opción no válida en el sistema.");
}
```

```java
import java.util.Scanner;

public class ClasificadorDatos {
    public static void main(String[] args) {
        Scanner scanner = new Scanner(System.in);
        
        System.out.println("--- SENSOR DE ALERTA AMBIENTAL ---");
        System.out.print("Ingrese la temperatura del sistema (double): ");
        double nivelTemp = scanner.nextDouble();
        
        System.out.print("Ingrese presión del cilindro (int): ");
        int presion = scanner.nextInt();
        
        // Evaluamos doble factor de riesgo
        if (nivelTemp > 80.0 && presion > 100) {
            System.out.println("¡ALERTA ROJA! Riesgo inminente de explosión térmica.");
        } else if (nivelTemp > 80.0 || presion > 100) {
            System.out.println("Aviso: Un valor excedió el límite de control. Por favor, revise de inmediato.");
        } else {
            System.out.println("Sistema en condiciones operativas óptimas. Riesgo calculado bajo.");
        }
        
        scanner.close();
    }
}
```

# Clase 4: Estructuras Repetitivas (Ciclos)

Continuando en nuestra ruta a través de los conceptos fundamentales, hoy nos enfocamos en el pilar de la automatización: el **loop** o **ciclo**. El corazón de la ciencia de datos implica poder entrenar modelos, analizar filas y leer enormes cantidades de datos utilizando estructuras repetitivas eficientes.

En Java, tenemos tres tipos de ciclos principales: `while`, `do-while`, y `for`. Cada uno atiende un caso de uso distinto.

## 1. Ciclo `while` (Mientras)
El ciclo `while` repite el bloque de código que contiene mientras(while) una condición evaluada permanezca como `true`. Su estructura es similar a la del ciclo `if`, pero en vez de evaluar la expresión lógica y ejecutarla una vez, ¡la evalúa contínuamente hasta que la expresión lógica devuelva `false`!
Se usa principalmente cuando **no sabemos exactamente cuántas iteraciones** vamos a realizar, pero sabemos la "condición de paro" (stop condition).

```java
int contador = 0; // Inicializamos un iterador

while (contador < 5) {
    System.out.println("Iteración número: " + contador);
    contador++; // contador++ es equivalente a decir: contador = contador + 1;
}
// El ciclo se detiene cuando el iterador es 5.
```

## 2. Ciclo `do-while` (Hacer-Mientras)
La gran diferencia con `while`, es que `do-while` **garantiza que el bloque de código se va a ejecutar al menos una vez**, ya que la condición lógica es evaluada sólo al final.

```java
import java.util.Scanner;
// ... (omitiendo public class y main para brevedad)
Scanner teclado = new Scanner(System.in);
int opcion;
// Es excelente para menús desde la terminal
do {
    System.out.println("Presiona 0 para salir en esta iteración. Otro número para continuar:");
    opcion = teclado.nextInt();
} while (opcion != 0); // Se repite MIENTRAS la opcion sea distinta de 0
```

## 3. Ciclo `for` (Para cada)
Es tu caballero fiel en Computación y Ciencia de Datos. Lo utilizas cuando **sabes exactamente** la cantidad de veces que debe realizarse la iteración, o bien para recorrer arrays de elementos.

El ciclo for concentra 3 elementos en la misma línea separados por `;`: `(inicialización; condición; incremento)`.

```java
// El ciclo For es la joya de la corona para operaciones en serie
for (int i = 0; i < 10; i++) {
    System.out.println("Procesando epoch n.º " + i);
    // 'i' será primero 0 y se ejecutará. 0 < 10, i aumenta a 1. Y así sucesivamente.
}
// La variable 'i' ya no existe fuera de estas llaves.
```

## 4. Ejemplos en Clase
Abre tu Apache NetBeans y copia este código completo. Es un simulador que utiliza el ciclo for para representar el crecimiento compuesto, muy usado para medir rendimientos e interés.

```java
import java.util.Scanner;

public class SimulacionCrecimiento {
    public static void main(String[] args) {
        Scanner input = new Scanner(System.in);
        
        System.out.print("Introduce la población inicial (ej: 100): ");
        double poblacion = input.nextDouble();
        
        System.out.print("Tasa de crecimiento por año en porcentaje (ej: 5 para 5%): ");
        double tasa = input.nextDouble() / 100.0; // Lo convertimos a fracción
        
        System.out.print("Cantidad de años a proyectar (ej: 10): ");
        int anios = input.nextInt();
        
        System.out.println("\n--- PROYECCIÓN INICIADA ---");
        
        // Ciclo For, vamos desde el año 1, hasta que alcancemos y procesemos el total solicitado
        for (int i = 1; i <= anios; i++) {
            poblacion = poblacion + (poblacion * tasa);
            // Mostramos los resultados iteración a iteración formatados para evitar tantos decimales
            System.out.printf("Año %d -> Población final calculada: %.2f\n", i, poblacion);
        }
        
        System.out.println("\nFin de la simulación. Datos proyectados exitosamente.");
        
        input.close();
    }
}
```

# Clase 5: Arreglos (Arrays) y Matrices

En el ámbito de la Ciencia de Datos, un solo dato es historia, muchos datos son estadística. Así como en matemáticas manipulamos desde escalares puros (variables de las clases 1 a 4) hasta vectores y matrices de múltiples dimensiones, en la práctica computacional pura de Java esto corresponde a los _Arreglos_ (`Arrays`).

Un Arreglo no es más que una caja conteniendo muchas divisiones del mismo tipo. Si creas la regla de una cajonera estricta que va guardar pantalones (`int`), esa cajonera no podría albergar guantes (`boolean`). 

## 1. Arrays Unidimensionales (Vectores)
En Java los Arrays son estáticos, lo que significa que **tienen un tamaño fijo**. Si declaro que un arreglo va a retener 5 elementos, no lo puedes expandir después sin esfuerzo adicional.

```java
// Forma 1: Cuando YA sabemos cuáles son los datos. Directo a memoria.
int[] puntuaciones_sat = {1200, 1500, 1050, 1340, 950};

// Forma 2: Cuando sabemos el TAMAÑO pero los datos vendrán vacíos y se insertarán por algoritmo
double[] probabilidadVentas = new double[4]; // Arreglo de longitud 4 lleno de ceros (0.0).

// Acceso a posiciones y mutación
// Importante en CS: ¡Los arreglos comienzan su índice iterativo desde cero a N-1 elementos!
probabilidadVentas[0] = 0.52; 
probabilidadVentas[1] = 0.70; 
// probabilidadVentas[4] = 0.11; // 💥 ¡ERROR FATAL! ArrayIndexOutOfBoundsException (Solo llega de 0 a 3)
```

## 2. Recorriendo Arreglos
La forma más canónica de procesar una carga de datos en arrays masivos es con bucles, particularmente el ciclo `for`. Todo arreglo trae una propiedad intrínseca `.length` (su longitud total) que te salva de intentar recordar su tamaño a mano.

```java
double sumaT = 0;
for (int i = 0; i < puntuaciones_sat.length; i++) {
   sumaT = sumaT + puntuaciones_sat[i];
}
System.out.println("Promedio general del SAT : " + (sumaT / puntuaciones_sat.length) );
```

## 3. Arrays Bidimensionales (Matrices)
¿Trabajas con imágenes digitales o en tabular (Tablas CSV/DataFrames/Hojas de cálculo)? Acabas de encontrar la Matriz. Los arreglos bidimensionales son "Listas de Listas" y poseen una relación espacial en un eje "X - Y" (`filas` y `columnas`). 

```java
// Filas y columnas (Matrix de 3 filas, 3 columnas)
int[][] matrizPlanaImagePixeles = {
    {255, 128,  0}, // Fila 0
    {10,  255, 15}, // Fila 1
    {50,   84, 99}  // Fila 2
};

// Acceso al renglon(Fila) 1, columna 0. Debería darnos 10.
System.out.println(matrizPlanaImagePixeles[1][0]);
```

## 4. Ejemplos en Clase
Abre tu NetBeans. En copia y prueba las proezas de recorrer un vector aleatorio calculando el valor Máximo y el Mínimo matemáticamente. Muy práctico y básico al explorar datasets.

```java
public class AnaliticaVectores {
    public static void main(String[] args) {
        // Dataset hipotético de temperaturas leídas en una semana
        double[] gradosRegistradosStr = {22.5, 23.1, 19.8, 17.0, 25.4, 21.3, 20.0};
        
        // Suposición inicial: asumimos en un principio el Máximo y el Mínimo va ser el primer elemento
        double tempMax = gradosRegistradosStr[0];
        double tempMin = gradosRegistradosStr[0];
        
        // Arrancamos desde 1 dado que la primer caja ya la comprobamos consigo mismo
        for(int i = 1; i < gradosRegistradosStr.length; i++) {
            if(gradosRegistradosStr[i] > tempMax) {
                tempMax = gradosRegistradosStr[i]; 
            }
            if(gradosRegistradosStr[i] < tempMin) {
                tempMin = gradosRegistradosStr[i];
            }
        }
        
        System.out.println("--- RESULTADOS ANALÍTICOS ---");
        System.out.println("Total de datos escaneados: " + gradosRegistradosStr.length);
        System.out.println("Temperatura Máxima detectada: " + tempMax);
        System.out.println("Temperatura Mínima detectada: " + tempMin);
    }
}
```

# Clase 6: Funciones, Métodos y Modularidad

Has llegado a la última clase.  En la ingeniería y la ciencia, enfrentar grandes problemas requiere a veces un enfoque puramente modular ("Divide y vencerás"). En Python, es habitual definir `def calcular_algo()`. En el territorio de Java a estas "funciones" les llamamos respetuosamente **Métodos** dado su íntima relación con las clases.

## 1. ¿Por qué utilizar Métodos?
Hasta ahora hemos escrito todo el código dentro de `public static void main`. Para las primeras cinco clases nos salvaba, pero un programa completo real tendría miles de líneas inmanejables si lo metes todo ahì.
**Beneficios:** 
* Código Reutilizable (llámalo tantas veces quieras en distintas parte evitando repetir el bloque entero).
* Facilita localizar "bugs" encapsulando lógicas aisladas.
* Organización y abstracción de problemas enormes.

## 2. Sintaxis de un Método
Los métodos se deben crear *fuera* de `main(...)` pero *siempre dentro* de las llaves envolventes de `public class`. 

> Un Método Java requiere especificar imperativamente su **Nivel de Privacidad/Alcance**, su **Retorno** y sus **Parámetros**.

```java
public class EjercicioMetodos {

    // 1er Método - Estructura básica de un "Procedimiento" que no devuelve respuesta sino que ejecuta acciones.
    // Al usar 'void' especificamos "Oye, haz lo tuyo y no esperamos que devuelvas ningún dato al terminar."
    public static void limpiarPipelineDeDatos() {
        System.out.println("Llamada ejecutada. Se han limpiado las caches temporales ETL.");
    }
    
    // 2do Método - "Función Pura". Devuelve o entrega un dato de regreso donde lo llamaron.
    // Usamos el tipo 'double' en vez de void y pasamos la palabra clave return.
    // Este método toma 2 argumentos a fuerza para poder vivir (a, y b)
    public static double calcularDistanciaMinkowski(double a, double b) {
        double dist = Math.abs(a - b);
        return dist; 
    }

    public static void main(String[] args) {
        // En main, es donde en realidad llamamos a estas acciones subalternas hechas a la medida.
        limpiarPipelineDeDatos();
        limpiarPipelineDeDatos(); // ¡Lo hemos reutilizado 2 veces con 1 línea!
        
        // Para funciones "return" puros guardamos o imprimimos su resultado al instante
        double metricaDistancia = calcularDistanciaMinkowski(170.3, 102.1);
        System.out.println("Distancia capturada por el sensor: " + metricaDistancia);
    }
}
```

## 3. Ejemplo Integrador Final
Combinaremos todo lo que hemos aprendido: Variables, Entradas (Scanner), Operadores Relacionales, y un bucle utilizando un Método estático hecho por nosotros. Es tu prueba de madurez ante Apache NetBeans.

```java
import java.util.Scanner;

public class CalculadoraEstadisticaDummies {

    // ===== SECCIÓN DE CREACIÓN DE UTILERÍAS MÓDULARES ===== 
    
    // Calcula la media aritmética de un vector completo de double.
    public static double resolverPromedioAritmetico(double[] serieDatos) {
        double sumaTotal = 0.0;
        for (int i = 0; i < serieDatos.length; i++) {
            sumaTotal += serieDatos[i];
        }
        return (sumaTotal / serieDatos.length);
    }
    
    // Método que evalua si hay anomalías basados rudimentariamente en si es negativo en este escenario inventado.
    public static boolean revisarFallasTensionEnElVector(double[] datosMecanica) {
        for(int x = 0; x < datosMecanica.length; x++) {
           if(datosMecanica[x] < 0) { 
               return true; // En el momento en que vemos falla rompemos el flujo y respondemos verdadero.
           }
        }
        return false; // Nunca encontramos un negativo, ergo no hay fallas, devolvemos false a base base principal 
    }

    // ===== FINAL MÉTODOS CREADOS =====

    public static void main(String[] args) {
        Scanner terminalIn = new Scanner(System.in);
        System.out.println("=== PREPARACIÓN DE ANALÍTICA ===");
        System.out.print("Cantidad de muestras obtenidas para el Vector: ");
        int longitudA = terminalIn.nextInt();
        
        // Iniciamos el vector dinámicamente basado de la terminal
        double[] arregloGeneral = new double[longitudA];
        
        System.out.println("--- INTRODUCE TUS VALORES (Puede incluir decimales. Ejemplo: 2.1) ---");
        
        for (int i = 0; i < arregloGeneral.length; i++) {
             System.out.print("Muestra para el índice [" + i + "]: ");
             arregloGeneral[i] = terminalIn.nextDouble();
        }
        
        // Llamando elegantemente a nuestros métodos encapsulados 
        double miPromedioCalculado = resolverPromedioAritmetico(arregloGeneral);
        boolean tenemosNegativosAlerta = revisarFallasTensionEnElVector(arregloGeneral);
        
        System.out.println("\n=== RESULTADO DEL MODELO ESTADÍSTICO ===");
        System.out.println("=> La media (promedio) es: " + miPromedioCalculado);
        
        if (tenemosNegativosAlerta) {
            System.out.println("=> RESULTADO ANÓMALO DE RESTRICCIÓN: Valores bajo cero detectados, sugiero descartar Dataset.");
        } else {
            System.out.println("=> RESULTADO DE RESTRICCIÓN: Todo operando en rango positivos normales.");
        }

        terminalIn.close();
    }
}
```
